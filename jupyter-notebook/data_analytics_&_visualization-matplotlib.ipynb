```python
from matplotlib import pyplot as plt
```

```python
# Simple Plots

x = [1, 5, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.show()
```

```python
# Multiple Subplots

plt.subplot(211)
plt.plot(x,y)
plt.subplot(212)
plt.plot(m,n)
plt.show()
```

```python
# Lets add title and lable to our graph

plt.plot(x, y)
plt.title('Mapinfo')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
m = [6, 3, 1]
n = [7, 8, 3]

plt.plot(m, n)
plt.show()
```

```python
# Adding Style to our graph

from matplotlib import style

plt.plot(x, y, 'red', label = 'line 1', linewidth = 5)
plt.plot(m, n, 'green', label = 'line 2', linewidth = 5)
plt.legend()

# You can also add the labels & tiles to the map

plt.grid(True, color = 'yellow')
plt.show()
```

```python
 # Bar Graph
    
plt.bar(x, y, label = 'example 1')
plt.bar(m, n, label = 'example 2', color = 'red')
plt.legend()

plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('info')

plt.show()
```

```python
# Histogram

import numpy as np

a = np.random.randn(1000)*15
b = np.arange(1,100, 2)

plt.hist(a, bins= b)
plt.show()
```

```python
# Scatter plot

p = np.random.rand(500)
q = np.random.rand(500)
#area = np.pi*3

# Plot
plt.scatter(p, q, s = 10, c = 'black')
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
# Pie Chart

slices = [7, 2, 2, 13]

activities = ['Arts', 'Science', 'Math', 'English']
color = ['red', 'blue', 'green', 'pink']

plt.pie(slices, 
       labels = activities, 
       colors = color,
       startangle = 90,
       shadow = True,
       explode = (0, 0.1, 0, 0),
       autopct = '%1.1f%%')

plt.title('Pie Plot')
plt.show()
```
