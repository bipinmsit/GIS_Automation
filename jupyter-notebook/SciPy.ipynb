```python
# Sub-package of SciPy

    - File input/output - scipy.io
    - Special Function - scipy.special
    - Linear Algebra Operation - scipy.linalg
    - Interpolation - scipy.interpolate
    - Optimization & Fit - scipy.optimize
    - Statistics and random numbers - scipy.stats
    - Numerical operation - scipy.integrate
    - Fast Fourier Transform - scipy.fftpack
    - Signal Processing - scipy.signal
    - Image Manipulation - scipy.ndimage
    - SciPy Spatial - scipy.spatial
    
    Note - SciPy is build on the top of NumPy
```

```python
import numpy as np
import scipy.io as sio
import scipy.special
import scipy.linalg
import scipy.fftpack
import scipy.optimize
import scipy.spatial
import scipy.integrate
import scipy.interpolate
import matplotlib.pyplot as plt
```

```python
# File input/output 

# Saves a MATLAB File - savemat
demo1 = np.arange(10)
sio.savemat('example.mat', {'demo':demo1})

# Loads a MATLAB File - loadmat
print(sio.loadmat('example.mat'))

print('\n')

# Lists variables inside a MATLAB File -whosmat
print(sio.whosmat('example.mat'))
```

```python
# Special Function

# Cubic Root Function
print(scipy.special.cbrt([8, 27, 64]))
print('\n')

print(scipy.special.exp10([1, 2, 3]))
print(scipy.special.exp2([2, 3, 4]))
print('\n')

# Permutations & Combinations - scipy.special.comb(N, k)/scipy.special.perm(N, k)
print(scipy.special.comb(5, 2, exact=True, repetition=False))
print(scipy.special.perm(5, 2, exact=True))
print('\n')

# Log Sum Exponential Function - scipy.special.logsumexp(x)
print(scipy.special.logsumexp([5, 10]))
print('\n')

# Gamma Function - scipy.special.gamma(x)
print(scipy.special.gamma([0, 1, 2, 3, 4, 5]))
```

```python
# Linear Algebra

# Solve the linear equation A*x + B*y = c

# x + y = 2
# x + 3y = 9

a = np.array([[1, 1], [1, 3]])
b = np.array([2, 9])

scipy.linalg.solve(a, b)
```

```python
# Linear Algebra

A = np.array([[1, 2], [3, 4]])

# Calculating determinanat - scipy.linalg.det()
print(scipy.linalg.det(A))
print('\n')

# Inverse Matrix - scipy.linalg.inv()
print(scipy.linalg.inv(A))
print('\n')

# Eigenvalues & Eigenvector - eg_value, eg_vect = scipy.linalg.eig()
print(scipy.linalg.eig(A))
print('\n')
```

```python
# Intergration (quad, dblquad, tplquad, nquad)

# integration of a function f with limit a = 0 & b = 1
f = lambda x: x**2

print(scipy.integrate.quad(f, 0, 1))
```

```python
# Discreate Fourier Transform


# Fast Fourier Transform - scipy.fftpack.fft()
x = np.array([1, 2, 1, -1, 1.5])
y = scipy.fftpack.fft(x)
z = scipy.fftpack.ifft(y)
print(y)

print('\n')

print(z)

# Discreate Cosine Transform - scipy.fftpack.dct()
```

```python
# Interpolation

x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(x**2/9.0)
f1 = scipy.interpolate.interp1d(x, y, kind='linear')
f2 = scipy.interpolate.interp1d(x, y, kind='cubic')


xnew = np.linspace(0, 10, num=41, endpoint=True)

plt.plot(x, y, 'o', xnew, f1(xnew), xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
# help(scipy.interpolate.interp1d)
# plt.plot(x, '*', color = 'red')
# plt.show()
```

```python
# SciPy - Spatial

# Delaunay Traingulations
scipy.spatial.Delaunay()

# Convex hulls
scipy.spatial.convex_hull_plot_2d()
```

```python
# Image Processing

# MISC Package in SciPy contains prebuilt images which can be used to perform image manipulation task

from scipy import misc
from scipy import ndimage
from matplotlib import pyplot as plt
import numpy as np

# get face image of panda from misc package
panda = misc.face()

# Show image of face
plt.imshow(panda)
plt.show()
```

```python
# Rotation of images

panda_roatate = ndimage.rotate(panda, 90)

plt.imshow(panda_roatate)
plt.show()
```
