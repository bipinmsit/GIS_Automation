```python
# Merging & Joining

df1 = pd.DataFrame({
                    'name':['Bipin', 'Nick', 'John'], 
                    'age':[20, 23, 22],
                    'city':['Bangalore', 'Vellore', 'Kolkata']})

df2 = pd.DataFrame({
                    'name':['Bipin', 'Nick', 'Rahul'], 
                    'age':[20, 25, 22],
                    'city':['Bangalore', 'Vellore', 'Kolkata']})

df1.merge(df2)
```

```python
df1.merge(df2, on='name')
```

```python
df1.merge(df2, on='age')
```

```python
# Joining
df1 = pd.DataFrame({
                    'int_rate': [2, 1, 2, 3],
                    'gdp': [50, 45, 45, 67]}, index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({
                    'HDI':[50, 45, 67, 34], 
                    'unemployment':[1, 3, 5, 6]}, index = [2001, 2003, 2004, 2004])
```

```python
df1.join(df2)
```
